# Conditionals

```python
x = 10
y = 15

if x < y:
    print("Passed test")
elif x == y:
    print("Passed test")
else:
    print("Failed test")

# Other options are != (not equals) or >

b1 = True
b2 = False

if b1 and b2:
    print("Passed test")

if b1 or b2:
    print("Passed test")

if not b1:
    print("Passed test")
```

```python
cTest = 'v'

match cTest:
    case 'c':
        print("First case")
    case 'v' | 'b':
        print("Second case")
    case 'n':
        print("Third case")
    case _:
        raise ValueError("Invalid input!")
```
