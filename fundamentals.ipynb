```python
x = 10
y = 15

if x < y:
    print("Passed test")
elif x == y:
    print("Passed test")
else:
    print("Failed test")

# Other options are != (not equals) or >

b1 = True
b2 = False

if b1 and b2:
    print("Passed test")

if b1 or b2:
    print("Passed test")

if not b1:
    print("Passed test")
```

```python
from datetime import datetime

def get_category(transaction):
    match transaction[0]:
        case "Tesco":
            return "Groceries"
        case "Netflix" | "Steam":
            return "Entertainment"
        case "British Gas":
            return "Utilities"
        case "Amazon":
            return "Shopping"
        case _:
            return "Other"

transactions = [
    ("Tesco", 3.75, datetime(2025, 11, 16)),
    ("Steam", 18.99, datetime(2025, 11, 28)),
    ("Tesco", 1.25, datetime(2025, 11, 29)),
    ("Netflix", 12.99, datetime(2025, 11, 30)),
    ("British Gas", 75.00, datetime(2025, 12, 1)),
    ("Amazon", 10.00, datetime(2025, 12, 2)),
    ("Tesco", 1.25, datetime(2025, 12, 3)),
    ("Netflix", 12.99, datetime(2025, 12, 4)),
    ("British Gas", 75.00, datetime(2025, 12, 5)),
    ("Amazon", 10.00, datetime(2025, 12, 6)),
    ("Tesco", 1.25, datetime(2025, 12, 7)),
]

for i, transaction in enumerate(transactions):
    print(f"{i}: {get_category(transaction)}")

# Exercise: Write a function that computes the total spending for a given category over a given month
```
